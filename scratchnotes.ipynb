```python
import pandas as pd
import json

data = json.load(open("games_sample.json", encoding="utf-8"))
df = pd.DataFrame(data["games"])

# flatten the nested dict column
details = pd.json_normalize(df["app_details"])

# combine back with main columns
df_flat = pd.concat([df.drop(columns=["app_details"]), details], axis=1)

for col in df_flat.columns:
    print(col)
```
